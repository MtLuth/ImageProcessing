```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("text112.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def display(img):
    dpi = 80
    height = img.shape[0]
    width = img.shape[1]
    fig_size = width/float(dpi), height/float(dpi)

    fig = plt.figure(figsize=fig_size)
    ax = fig.add_axes([0,0,1,1])

    ax.axis('off')

    ax.imshow(img, cmap="gray")
    plt.show()
```

```python
display(image)
```

```python
inverted = cv2.bitwise_not(image, image)
display(inverted)
```

```python
scale_img = cv2.resize(inverted, None, fx=5, fy=5, interpolation=cv2.INTER_CUBIC)
```

```python
display(scale_img)
```

```python
gray_scale = cv2.cvtColor(scale_img, cv2.COLOR_RGB2GRAY)
display(gray_scale)
```

```python
thresh, im_bw = cv2.threshold(gray_scale, 100, 100, cv2.THRESH_BINARY)
display(img=im_bw)
```
